Ποιο είναι το correlation που προκύπτει στις δύο περιπτώσεις; Είναι ικανοποιητική η γραμμική εξάρτηση των δύο μεταβλητών; Για το ερώτημα αυτό να μελετήσετε την πηγή <a href="https://en.wikipedia.org/wiki/Correlation_and_dependence">εδώ</a>.</li>

<p align="justify">Να θεωρήσετε ότι το test set είναι το 20% του αρχικού dataset. Τι παρατηρείτε ότι συμβαίνει στην περίπτωση που δε γίνει κανονικοποίηση των δεδομένων; Πειραματιστείτε με τις τιμές του ρυθμού μάθησης και τον αριθμό των εποχών για να δείτε πώς αλλάζει το μέσο τετραγωνικό σφάλμα.<p>

```
!pip install numpy
!pip install matplotlib
!pip install pandas

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from scipy.stats import pearsonr
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

## Scratch Functions

```
def minmax(x):
  min_val = np.min(x)
  max_val = np.max(x)
  x = (x-min_val) / (max_val-min_val)
  return x

def calcMSE(x,y,b0,b1):
  error=0
  x=np.array(x)
  y=np.array(y)
  for i in range(len(x)):
    pred=b0+b1*x[i]
    error=error+(y[i]-pred)**2
  return (error/i)


def scratchSGD(x,y,epochs,lr_rate):
  x=np.array(x)
  y=np.array(y)
  b0=1
  b1=0
  print("Initial MSE: ",calcMSE(x,y,b0,b1),"\n")
  for i in range(epochs):
    
    x, y = shuffle(x,y)
    
    for j in range(len(x)):
      
      ypred=b0+b1*x[j]
      
      gradb0=-2*(y[j]-ypred)
      gradb1=-2*x[j]*(y[j]-ypred)
      
      b0=b0-lr_rate*gradb0
      b1=b1-lr_rate*gradb1
    
    print("Epoch: ",i+1," MSE: ",calcMSE(x,y,b0,b1))
  
  return b0,b1
```

## Dataset A

```
datasa = pd.read_csv("https://raw.githubusercontent.com/nkostopoulos/StochasticsLabPublic/master/lab1/data1a.csv")
print(datasa)
```

```
y1=datasa['views']
x1=datasa['likes']
```

```
corr1,_ = pearsonr(y1,x1)
print("Pearson Correlation: ", corr1)
```

Παρατηρούμε ότι υπάρχει κάποια συσχέτιση όχι ιδιαίτερα έντονη αλλά ικανοποιητική για να εφαρμόσουμε τον αλγόριθμο της γραμμικής παλινδρόμησης.

```
plt.scatter(x1, y1)
plt.show()
```

Παρατηρούμε ότι μπορεί να εντοπιστεί μια συσχέτιση καθώς για μεγάλο x υπάρχει η τάση να αυξάνεται το y, παρόλα αυτά δεν είναι ιδιαίτερα έντονη.

Στο επόμενο βήμα πρώτα χωρίζουμε το dataset μας σε train/test και ύστερα πραγματοποιούμε min-max normalization καθώς δεν επιθυμούμε να εισάγουμε πληροφορία στο test σετ μας.

```
split_prcnt1=0.2

X_train1, X_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size=split_prcnt1)

X_train1_norm=minmax(X_train1)
X_test1_norm=minmax(X_test1)

y_train1_norm=minmax(y_train1)
y_test1_norm=minmax(y_test1)
```

##### Non Normalized

```
b01,b11=scratchSGD(X_train1,y_train1,epochs=100,lr_rate=1)
print("MSE on TEST: ",calcMSE(X_test1,y_test1,b0=b01,b1=b11))
```

```
b01,b11=scratchSGD(X_train1,y_train1,epochs=100,lr_rate=0.9)
print("MSE on TEST: ",calcMSE(X_test1,y_test1,b0=b01,b1=b11))
```

Όταν δεν πραγματοποιούμε κανονικοποίηση στο dataset μας παρατηρούμε δύο φαινόμενα:


*   Ταλάντωση, δηλαδή όχι σύγκλιση, και αυτό το φαινόμενο παρατηρείται όταν έχουμε αρκετά μεγάλο learning rate
*   Υπερχείλιση, λόγω των τεράστιων τετραγωνικών διαφορών που προκύπτουν, η python δεν μπορεί να ανταπεξέλθει και έχουμε σαν ένδειξη το nan


##### Normalized

```
b01,b11=scratchSGD(X_train1_norm,y_train1_norm,epochs=5,lr_rate=0.9)
print("MSE on TEST: ",calcMSE(X_test1_norm,y_test1_norm,b0=b01,b1=b11))
```

```
b01,b11=scratchSGD(X_train1_norm,y_train1_norm,epochs=50,lr_rate=0.9)
print("MSE on TEST: ",calcMSE(X_test1_norm,y_test1_norm,b0=b01,b1=b11))
```

```
b01,b11=scratchSGD(X_train1_norm,y_train1_norm,epochs=50,lr_rate=0.01)
print("MSE on TEST: ",calcMSE(X_test1_norm,y_test1_norm,b0=b01,b1=b11))
```

```
b01,b11=scratchSGD(X_train1_norm,y_train1_norm,epochs=1000,lr_rate=0.00008)
print("MSE on TEST: ",calcMSE(X_test1_norm,y_test1_norm,b0=b01,b1=b11))
```

####Από τα παραπάνω πειράματα καταλήγουμε στο συμπέρασμα ότι αν αποφασίσουμε να έχουμε έναν χαμηλό αριθμό εποχών, τότε θα πρέπει να έχουμε και ένα σχετικά υψηλό learning rate. Σε αυτή την περίπτωση θα έχουμε κερδίσει σε χρονική απόδοση και θα έχουμε ένα καλό τελικό μοντέλο.

####Αν είμαστε διατεθειμένοι να αυξήσουμε τις εποχές με αποτέλεσμα να χάσουμε σε χρονική απόδοση, τότε θα πρέπει να χρησιμοποιήσουμε μικρό learning rate. Σε αυτή την περίπτωση είναι αρκετά πιθανό να έχουμε πετύχει καλύτερο μοντέλο συγκριτικά με την προηγούμενη περίπτωση.

#### Υπάρχει ένα plateau στο οποίο δεν μπορούμε να βελτιώσουμε παραπάνω την απόδοση του μοντέλου και αντιθέτως μειώνουμε την χρονική απόδοση στην εκπαίδευση του.

## Απαντήσεις

1. Οι κλάσεις που δημιουργούνται είναι οι : {1,2} η οποία είναι κλειστή , {3,4,5} η οποία είναι ανοιχτή και η {6,7,8} η οποία είναι κλειστή. Στο μεταβατικό στάδιο τοποθετούνται οι καταστάσεις που ανήκουν σε ανοιχτές κλάσεις , δηλαδή οι 3,4,5 είναι παροδικές ενώ οι 1,2 και 6,7,8 είναι επαναληπτικές.

2. Στην περίπτωση που η αλυσίδα ξεκινάει από την κατάσταση 1 θα χρησιμοποιήσω τον υποπίνακα $$\begin{bmatrix} 1/2 & 1/2 \\ 1/4 & 3/4 \end{bmatrix}$$ και λύνοντας το σύστημα π=πP και Σ π = 1, καταλήγω σε π=(1/3,2/3).

Για την περίπτωση που η αλυσίδα ξεκινάει από την κατάσταση 8 θα χρησιμοποιήσω τον υποπίνακα $$\begin{bmatrix} 0 & 1/2 & 1/2 \\ 1/2 & 0 & 1/2 \\ 1/2 & 1/2 & 0 \end{bmatrix}$$ για τον οποίο βλέπω ότι είναι διπλά στοχαστικός επομένως μπορώ αμέσως να πω ότι π = (1/3,1/3,1/3).

3. Ο αναμενόμενος αριθμός βημάτων υπολογίζεται ως 1/π2 = 3/2 , δηλαδή θα χρειαστούν 2 βήματα.
